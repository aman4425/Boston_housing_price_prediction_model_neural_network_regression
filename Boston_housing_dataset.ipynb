```python
import tensorflow as tf
print(tf.__version__)
```

```python
import pandas as pd
import numpy as np
```

```python
(x_train , y_train) , (x_test , y_test) = tf.keras.datasets.boston_housing.load_data(
    path='boston_housing.npz', test_split=0.2, seed=113
)
```

```python
x_train.shape  , y_train.shape
```

```python
x_test.shape , y_test.shape
```

### Creating the basic model

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

model.fit(x_train , y_train , epochs = 100)
```

```python
model.evaluate(x_test,y_test)
```

### Adding extra hidden layer with extra 10 hidden units and ADAM optimizer

```python
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(),
              metrics = ["mae"])

history = model_2.fit(x_train , y_train , epochs = 100)
```

```python
import matplotlib.pyplot as plt
pd.DataFrame(history.history).plot()
plt.xlabel("epochs")
plt.ylabel("loss")
```

```python
model_2.evaluate(x_test,y_test)
```

### Adding one more layer to check model

```python
tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_3.compile(loss = tf.keras.losses.mae,
                optimizer=tf.keras.optimizers.Adam(),
                metrics = ["mae"])

model_3.fit(x_train,y_train, epochs=100)
```

```python
model_3.evaluate(x_test, y_test)
```

### now checking model 2 with learning rate 0.01 to check the accuracy of model 2

```python
tf.random.set_seed(42)

model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_4.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics = ["mae"])

model_4.fit(x_train , y_train , epochs = 100)
```

```python
model_4.evaluate(x_test, y_test)
```
